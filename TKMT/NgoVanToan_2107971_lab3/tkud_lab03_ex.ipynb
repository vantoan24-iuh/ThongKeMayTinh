## NUMPY

Bạn hãy phát sinh ngẫu nhiên mảng ndarray có tên `arr` gồm 100 phần tử số nguyên có giá trị từ 0..1000. Thực hiện các thao tác sau:

````python
import numpy as np
from numpy import random
````

````python
arr=random.randint(1000,size=100)
print(arr)
````

### 1. Hiển thị các phần tử thỏa yêu cầu dưới đây

- 10 phần tử đầu tiên
- 5 phần tử cuối cùng
- Các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100

````python
# - 10 phần tử đầu tiên
print("10 phan tu dau tien la: ",arr[:10])
# - 5 phần tử cuối cùng
print("5 phan tu cuoi cung la: ",arr[-5:])
# - Các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100
for i in arr:
    if (i % 5 ==0 and i>100):
        print(i)
````

### 2. Cho biết mảng trên có bao nhiêu phần tử có giá trị nhỏ hơn 100. Hiển thị các phần tử đó

````python
for i in arr:
    if(i<100):
        print(i)
````

### 3. Thực hiện thống kê:

- Tính tổng mảng
- Tìm giá trị trung bình
- Tìm giá trị trung vị
- Tìm Q1, Q2, Q3
- Tìm phân vị 20 (percentile 20)

````python
# Tinh tong mang
print(np.sum(arr))
# Tinh gia tri trung binh
print(np.average(arr))
# Tinh gia tri trung vi
print(np.median(arr))
# Timg Q1,Q2,Q3
print("Q1 = ",np.quantile(arr,0.25))
print("Q1 = ",np.quantile(arr,0.5))
print("Q1 = ",np.quantile(arr,0.75))
# Tim phan vi 20 (percentile 20)
print("Phan vi 20 = ",np.quantile(arr,0.2))
````

### 4. Thực hiện việc chọn ngẫu nhiên (không thay thế) 10 phần tử từ mảng trên 5 lần. Hiển thị kết quả lựa chọn mỗi lần

````python
print(random.choice(arr,size=10,replace=False))
print(random.choice(arr,size=10,replace=False))
print(random.choice(arr,size=10,replace=False))
print(random.choice(arr,size=10,replace=False))
print(random.choice(arr,size=10,replace=False))
````

### 5. Tìm phần tử xuất hiện nhiều nhất trong mảng

````python
most_common = np.bincount(arr).argmax()
print("Phần tử xuất hiện nhiều nhất trong mảng:", most_common)
````

### Câu 4: Định dạng đường thẳng

Phát sinh x là các điểm dữ liệu từ $[0, \pi]$. Bạn hãy vẽ đồ thị $y = sin(x)$ và $y = cos(x)$ như dưới đây.
Gợi ý: sử dụng hằng số pi và hàm sau:
- `np.pi` : số $\pi$
- `np.linspace()`
- `np.sin()`
- `np.cos()`

````python
x = np.linspace(0,6,1000)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,label='sin(x)')
plt.plot(x,z,label='cos(x)')
plt.grid(True)
plt.legend() #hiển thị chú thích trên biểu đồ
````

### Câu 5: Vẽ các đồ thị con

````python
x = np.linspace(0,6, 1000)
fig, ax = plt.subplots(2,1, constrained_layout=True)
ax[0].plot(x, np.sin(x), label='sin(x)')
ax[1].plot(x, np.cos(x), label='cos(x)', ls='dashed', c='red')

ax[0].legend();
ax[1].legend();
````

### Câu 8: Vẽ histogram

Sử dụng hàm `np.random.normal(size=10000)` để phát sinh 10000 số ngẫu nghiên thuộc phân phối chuẩn. Vẽ histogram các số ngẫu nhiên trên.

### Câu 9: Vẽ đồ thị hộp

````python
x = np.random.normal(size=10000)
plt.boxplot(x)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
````

### Câu 10: Kết hợp trên cùng một đồ thị

Sử dụng hàm `np.random.normal(size=10000)` để phát sinh 10000 số ngẫu nghiên thuộc phân phối chuẩn. Vẽ hình thể hiện phân phối như sau:

Hướng dẫn: thêm tham số `vert=False` trong hàm vẽ `boxplot()`

````python
x = np.random.normal(size=10000)
fig, ax = plt.subplots(2,1, constrained_layout=True)
ax[0].hist(x, bins=100, density=True, alpha=0.7)
ax[1].boxplot(x,vert=False)
ax[0].set_title('Histogram')
ax[1].set_title('Boxplot')
fig.suptitle('Đồ Thị Histogram và Boxplot')
````

## SEABORN



Seaborn là một thư viện hỗ trợ trực quan hóa được viết dựa trên thư viện `matplotlib`. Trong bài tập này, bạn hãy tự tìm hiểu thư viện `seaborn` và trực quan hóa dữ liệu tiền tip tại một nhà hàng.

Bạn có thể dùng đoạn mã sau để đọc dữ liệu có sẵn trong `seaborn`

```python
import seaborn as sns

df = sns.load_dataset("tips")
df.head()
```
Hoặc bạn có thể đọc dữ liệu chứa trong file `tips.csv` chứa thông tin tiền tip của khách hàng tại một nhà hàng. Dữ liệu có các thông tin sau:

- `total_bill`: tổng hoá đơn (USD)
- `tip`: tiền tip/ tiền boa (USD)
- `sex`: giới tính người thanh toán hóa đơn
- `smoker`: có hút thuốc hay không
- `day`: ngày trong tuần
- `time`: thời gian dùng bữa(buổi sáng, buổi trưa, buổi tối)
- `size`: số lượng người tham gia

Bạn hãy lựa chọn loại đồ thị phù hợp để trả lời những câu hỏi sau:

````python
import seaborn as sns
tip= sns.load_dataset("tips")
tip.head()
````

### 1. Vẽ đồ thị thể hiện tổng tiền tip theo ngày

````python
total_tip = tip.groupby(by="day")["tip"].sum()
sns.barplot(x=total_tip.index,y=total_tip, data=tip, errorbar=None)
plt.show();
````

### 2. Vẽ đồ thị thể hiện tổng tiền tip theo thời gian dùng bữa (time)


````python
total_tip = tip.groupby(by="time")["tip"].sum()
sns.barplot(x=total_tip.index, y=total_tip, data=tip, errorbar=None)
plt.show();
````

### 3. Vẽ đồ thị histogram của tổng hoá đơn


````python
sns.histplot(data=tip,x='total_bill', bins=50)
plt.show();
````

### 4. Vẽ đồ thị boxplot của tổng hoá đơn


````python
sns.boxplot(data=tip ,x='total_bill')
plt.show();
````

### 5. Vẽ đồ thị boxplot so sánh tổng hóa đơn giữa các ngày trong tuần


````python
sns.boxplot(data=tip ,x='day', y='total_bill', hue='day')
plt.show();
````

### 6. Vẽ đồ thị để kiểm tra sự liên quan giữa tổng hoá đơn và tiền tip. (`scatter plot`)


````python
sns.scatterplot(data=tip,x='total_bill', y='tip', hue='total_bill')
plt.show();
````

### 7. Tìm hiểu và sử dụng hàm `jointplot` trong seaborn để khảo sát quan hệ giữa hai thuộc tính `total_bill` và `tip`. Theo bạn hai thuộc tính này có tương quan với nhau hay không?


````python
sns.jointplot(data=tip, x='total_bill', y='tip')

plt.show();
````

### 8. Tìm hiểu và sử dụng hàm `pairplot` trong seaborn để trực quan bộ dữ liệu. Theo bạn, trong bộ dữ liệu có những thuộc tính nào liên quan với nhau không?



````python
sns.pairplot(data=tip)
plt.show();
````

### 9. Vẽ hai đồ thị histogram cạnh nhau để so sánh tổng hóa đơn giữa hai thời gian dùng bữa `Lunch` và `Dinner`



````python
fig, ax = plt.subplots(1,2, constrained_layout=True)
sns.histplot(data=tip[tip['time']=='Lunch'], x='total_bill', bins=50, ax=ax[0])
ax[0].set_title('Lunch')
sns.histplot(data=tip[tip['time']=='Dinner'], x='total_bill', bins=50, ax=ax[1])
ax[1].set_title('Dinner')
fig.suptitle('So sánh tổng hóa đơn giữa hai thời gian dùng bữa')
plt.show();
````

### 10. Vẽ hai đồ thị boxplot cạnh nhau để so sánh tổng hóa đơn giữa hai thời gian dùng bữa `Lunch` và `Dinner`

````python
fig, ax = plt.subplots(1,2, constrained_layout=True)
sns.boxplot(data=tip[tip['time']=='Lunch'], x='time', y='total_bill', ax=ax[0])
ax[0].set_title('Lunch')
sns.boxplot(data=tip[tip['time']=='Dinner'], x='time', y='total_bill', ax=ax[1])
ax[1].set_title('Dinner')
fig.suptitle('So sánh tổng hóa đơn giữa hai thời gian dùng bữa')
plt.show();
````
