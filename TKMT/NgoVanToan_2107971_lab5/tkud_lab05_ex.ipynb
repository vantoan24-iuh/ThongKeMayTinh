### Bài 1: Tạo quần thể mô phỏng

Bạn hãy tạo một mẫu gồm 10000 phần tử mô phỏng chiều cao của nam thanh niên có giá trị từ 120cm - 200cm. Bạn hãy lưu kết quả vào biến **POP**.

Tính:
a. Chiều cao trung bình (kỳ vọng) của quần thể và độ lệch chuẩn về chiều cao của quần thể.

b. Tính tỷ lệ người cao trong quần thể, biết rằng thanh niên có chiều cao từ 180 trở lên được gọi là cao.

c. Vẽ histogram về chiều cao của quần thể. Theo bạn quần thể có phân phối chuẩn hay không?

d. Thử vẽ histogram và đồ thị hàm mật độ của phân phối chuẩn sử dụng tham số loc và scale bằng với kỳ vọng và độ lệch chuẩn của quần thể.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# a. Chiều cao trung bình (kỳ vọng) của quần thể và độ lệch chuẩn về chiều cao của quần thể.
POP = np.random.randint(120, 201, 10000)
mean = np.mean(POP)
std = np.std(POP)
print(f'Chiều cao trung bình của quần thể: {mean}')
print(f'Độ lệch chuẩn về chiều cao của quần thể: {std}')

# b. Tính tỷ lệ người cao trong quần thể, biết rằng thanh niên có chiều cao từ 180 trở lên được gọi là cao.
tyle_cao = len(POP[POP >= 180]) / len(POP)
print(f'Tỷ lệ người cao trong quần thể: {tyle_cao}')

# c. Vẽ histogram về chiều cao của quần thể. Theo bạn quần thể có phân phối chuẩn hay không?
sns.histplot(POP, kde=True)
plt.show()
# Theo biểu đồ, quần thể không có phân phối chuẩn.

# d. Thử vẽ histogram và đồ thị hàm mật độ của phân phối chuẩn sử dụng tham số loc và scale bằng với kỳ vọng và độ lệch chuẩn của quần thể
sns.histplot(POP, kde=True)
x = np.linspace(POP.min(), POP.max(), 1000)
y = 1/(std * np.sqrt(2 * np.pi)) * np.exp(- (x - mean)**2 / (2 * std**2))
plt.plot(x, y, 'r')
plt.show()
# Kết quả cho thấy hàm mật độ của phân phối chuẩn khá giống với histogram của quần thể.

# Nhận xét: Quần thể không có phân phối chuẩn, nhưng hàm mật độ của phân phối chuẩn khá giống với histogram của quần thể.
```

### Bài 4: Khảo sát phân phối của trung bình mẫu

a. Thực hiện bài 2 100 lần, mỗi lần bạn tính được trung bình mẫu. Vẽ đồ thị histogram cho 100 trung bình mẫu bạn tính được.

```python
# Khảo sát phân phối của trung bình mẫu

# a. Thực hiện bài 2 100 lần, mỗi lần bạn tính được trung bình mẫu. Vẽ đồ thị histogram cho 100 trung bình mẫu bạn tính được.
means = []
for i in range(100):
    SAMPLE = np.random.choice(POP, 100)
    mean_sample = np.mean(SAMPLE)
    means.append(mean_sample)
sns.histplot(means, kde=True)
plt.show()


# b. Thực hiện bài 3 100 lần, mỗi lần bạn tính được trung bình mẫu. Vẽ đồ thị histogram cho 100 trung bình mẫu bạn tính được.
means = []
for i in range(100):
    SAMPLE = np.random.choice(POP, 20)
    mean_sample = np.mean(SAMPLE)
    means.append(mean_sample)
sns.histplot(means, kde=True)
plt.show()

# c. Vẽ histogram của hai câu a, b trong cùng một hình để so sánh. Theo bạn hình dạng của histogram thay đổi nói lên điều gì?
means1 = []
means2 = []
for i in range(100):
    SAMPLE = np.random.choice(POP, 100)
    mean_sample = np.mean(SAMPLE)
    means1.append(mean_sample)
    SAMPLE = np.random.choice(POP, 20)
    mean_sample = np.mean(SAMPLE)
    means2.append(mean_sample)
sns.histplot(means1, kde=True, color='red', alpha=0.5)
sns.histplot(means2, kde=True, color='blue', alpha=0.5)
plt.show()
# Gợi ý: tạo một danh sách rỗng để chứa các trung bình mẫu tính được, sử dụng hàm append() để thêm giá trị trung bình vào danh sách sau mỗi lần tính.
```

### Bài 7:  Bài Tập Ước Lượng 

Khảo sát cân nặng (kg) của gà khi xuất chuồng, người ta cân một số con và có kết quả như sau:

> 2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8

Giả sử cân nặng của gà là biến ngẫu nhiên có phân phối chuẩn. Hãy lập khoảng tin cậy cho cân nặng trung bình của  khi xuất chuồng với độ tin cậy là 95%. Trong 2 trường hợp:

a, Biết $\sigma = 0.3$

b, Không biết $\sigma$


```python
data = np.array([2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8])
# Giả sử cân nặng của gà là biến ngẫu nhiên có phân phối chuẩn. Hãy lập khoảng tin cậy cho cân nặng trung bình của  khi xuất chuồng với độ tin cậy là 95%. Trong 2 trường hợp: phương sai của quần thể là 0.3 và phương sai của quần thể là phương sai mẫu.
mean = np.mean(data)
std = np.std(data)
n = len(data)
alpha = 0.05
z = 1.96
# Phương sai của quần thể
var = 0.3
lower = mean - z * np.sqrt(var / n)
upper = mean + z * np.sqrt(var / n)
print(f'Khoảng tin cậy cho cân nặng trung bình của gà khi phương sai của quần thể là 0.3: ({lower}, {upper})')
# Phương sai mẫu
var = np.var(data)
lower = mean - z * np.sqrt(var / n)
upper = mean + z * np.sqrt(var / n)
print(f'Khoảng tin cậy cho cân nặng trung bình của gà khi phương sai của quần thể là phương sai mẫu: ({lower}, {upper})')
```
