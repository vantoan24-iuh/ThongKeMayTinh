a, (1 điểm) Vẽ đồ thị histogram dân số các nước năm 2000. (Gợi ý: sử dụng hàm `np.log` để vẽ histogram của thuộc tính population thay vì vẽ histogram trực tiệp

```python
# Vẽ đồ thị histogram dân số các nước năm 2000. (Gợi ý: sử dụng hàm `np.log` 
# để vẽ histogram của thuộc tính population thay vì vẽ histogram trực tiệp
population_2000=np.log(df_country['2000'])
plt.hist(population_2000,bins=100)
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title('Population distribution in 2000')
plt.show()
```

```python
# Vẽ đồ thị boxplot dân số các nước từ năm 2010 - 2020 (gợi ý: sử dụng hàm `melt()`)
population_2010_2020=pd.melt(df_country,id_vars=['name'],value_vars=['2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020'],var_name='year',value_name='population') 
plt.figure(figsize=(10,7))
sns.boxplot(x='year',y='population',data=population_2010_2020)
plt.show()
```

---

```python
# Vẽ biểu đồ tăng trưởng dân số của 10 quốc gia Đông Nam Á từ năm 2000-2020
df_country_10asia_2000_2020=pd.melt(df_Country_asia,id_vars=['name'],value_vars=['2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020'],var_name='year',value_name='population')
plt.figure(figsize=(15,7))
sns.lineplot(x='year',y='population',hue='name',data=df_country_10asia_2000_2020)
plt.show()
```

 b, (0.5 điểm) Dựa vào hình vẽ bạn hãy cho biết quốc gia nào có tốc độ tăng trưởng nhanh nhất? 

Nhìn vào đồ thị thấy được nước có tăng trưởng cao nhất là Indonesia với tốc độ khá nhanh

```python
# Vẽ biểu đồ tăng trưởng về dân số của 10 quốc gia Đông Nam Á từ năm 2000-2020
gmp_ctrs_sea = df_country[df_country['name'].isin(['Vietnam', 'Thailand', 'Indonesia', 'Philippines', 'Malaysia', 'Singapore', 'Myanmar', 'Cambodia', 'Laos', 'Brunei'])]
gmp_ctrs_sea = gmp_ctrs_sea.melt(id_vars=['geo', 'name'], value_vars=['2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020'], var_name='Year', value_name='Population')

plt.figure(figsize=(10, 6))
# sns.lineplot(data=gmp_ctrs_sea, x='Year', y='Population', hue='name', ci=None, marker="o")
for country in gmp_ctrs_sea['name'].unique():
    plt.plot(gmp_ctrs_sea[gmp_ctrs_sea['name'] == country]['Year'], gmp_ctrs_sea[gmp_ctrs_sea['name'] == country]['Population'], marker='o', label=country)
plt.ylabel('Population Growth Rate')
plt.title('Population Growth Rate of 10 Southeast Asia Countries from 2000 to 2020')
plt.xticks(gmp_ctrs_sea['Year'], rotation=45)
plt.grid()
plt.legend()
plt.show()
```
