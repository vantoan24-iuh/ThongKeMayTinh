### Câu 10 (1 điểm)

### Câu 11 (1 điểm)
Tìm bang có số lượng bé gái sinh ra nhiều nhất trong khoảng thời gian từ 2004 đến 2011

```python
Gender_Female = df_baby[df_baby['Gender']=='F'] # Lọc ra tất cả các dòng thuộc giới tính nữ
sort_Female = Gender_Female.sort_values(by='Count', ascending=False)
year_2004_2011=sort_Female[(sort_Female['Year'] >= 2010) & (sort_Female['Year'] <= 2011)]
top_1=year_2004_2011.head(1)
print(top_1['State'])
```

### Câu 12 (1 điểm)
Vẽ đồ thị cột (bar chart) top 10 tên xuất hiện nhiều nhất năm 2010 ở bang California (mã bang là "CA"). Đặt title của đồ thị là "Top 10 names in California in 2010".

```python
year_2010=df_baby[(df_baby['Year'] == 2010)]
states_CA=year_2010[(year_2010['State'] == 'CA')]
top_10_name=states_CA['Name'].head(10)
print(top_10_name)
plt.figure(figsize=(10,5))
plt.bar(top_10_name, states_CA['Count'][:10])
plt.title('Top 10 names in California in 2010')
plt.xlabel('Name')
plt.ylabel('Count')
plt.show()
```
