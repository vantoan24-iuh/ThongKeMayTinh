### Cu 10 (1 im)
Top 5 tn b trai c s dng nhiu nht?

```python
Gender_Male = df_baby[df_baby['Gender']=='M'] # Lc ra tt c cc dng thuc gii tnh nam
sort_name = Gender_Male.sort_values(by='Count', ascending=False) # Sp xp gim dn theo ct Count
top_5 = sort_name.head(5) # Ly ra top 5 tn
print(top_5['Name']) 
```

### Cu 11 (1 im)
Tm bang c s lng b gi sinh ra nhiu nht trong khong thi gian t 2004 n 2011

```python
Gender_Female = df_baby[df_baby['Gender']=='F'] # Lc ra tt c cc dng thuc gii tnh n
sort_Female = Gender_Female.sort_values(by='Count', ascending=False)
year_2004_2011=sort_Female[(sort_Female['Year'] >= 2010) & (sort_Female['Year'] <= 2011)]
top_1=year_2004_2011.head(1)
print(top_1['State'])
```

### Cu 12 (1 im)
V  th ct (bar chart) top 10 tn xut hin nhiu nht nm 2010  bang California (m bang l "CA"). t title ca  th l "Top 10 names in California in 2010".

```python
year_2010=df_baby[(df_baby['Year'] == 2010)]
states_CA=year_2010[(year_2010['State'] == 'CA')]
top_10_name=states_CA['Name'].head(10)
print(top_10_name)
plt.figure(figsize=(10,5))
plt.bar(top_10_name, states_CA['Count'][:10])
plt.title('Top 10 names in California in 2010')
plt.xlabel('Name')
plt.ylabel('Count')
plt.show()
```
