**Bài 02:** Một trung tâm bưu điện nhận được 3 cuộc gọi điện thoại mỗi phút. Tính xác suất để trung tâm nhận được 1 cuộc gọi, 2 cuộc gọi, 3 cuộc gọi trong vòng một phút biết số cuộc gọi trong một phút có phân phối Poisson.

```python
x = 1
lam = 3 
px = poisson.pmf(x, lam) # pmf là hàm mật độ xác suất, dùng để tính xác suất của một biến rời rạc

print('P(X=1):', px)

x = 2
px = poisson.pmf(x, lam)
print('P(X=2):', px)

x = 3
px = poisson.pmf(x, lam)
print('P(X=3):', px)
```

d. Từ hai cách lấy mẫu ở câu b, c bạn có nhận xét gì không?

```python
mu = 500
sigma = 4

# Sample sizes
sample_size_20 = 20
sample_size_100 = 100

# Generate random samples
weights_20 = np.random.normal(mu, sigma, sample_size_20)
weights_100 = np.random.normal(mu, sigma, sample_size_100)

# Plot histograms
plt.figure(figsize=(12, 4))

plt.subplot(121)
plt.hist(weights_20)
plt.xlabel("Trọng lượng (gam)")
plt.ylabel("Tần suất")
plt.title("Histogram trọng lượng trái cây (n = 20)")

plt.subplot(122)
plt.hist(weights_100)
plt.xlabel("Trọng lượng (gam)")
plt.ylabel("Tần suất")
plt.title("Histogram trọng lượng trái cây (n = 100)")
plt.show()

# Calculate mean and standard deviation for each sample
mean_20 = np.mean(weights_20)
std_20 = np.std(weights_20)
print("Giá trị trung bình (n = 20):", mean_20)
print("Độ lệch chuẩn (n = 20):", std_20)

mean_100 = np.mean(weights_100)
std_100 = np.std(weights_100)
print("Giá trị trung bình (n = 100):", mean_100)
print("Độ lệch chuẩn (n = 100):", std_100)
```

**Bài 04:** Mô phỏng tung một con xúc sắc cân đối đồng chất 5000 lần. Dựa vào giá trị mô phỏng, bạn hãy tìm các giá trị xác suất dưới đây:

a. Xác suất để  số chấm xuất hiện là 4

b. Xác suất để số chấm xuất hiện lớn hơn hoặc bằng 4

c. Giả sử biết số chấm xuất hiện lớn hơn hoặc bằng 4. Hãy tìm xác suất để mặt 6 chấm xuất hiện.

```python
# a. Xác suất để  số chấm xuất hiện là 4
num_rolls=5000
rolls=np.random.randint(1,7,num_rolls)
rolls4=np.count_nonzero(rolls==4)/num_rolls
print("Xac suat ra mat 4 la: ",rolls4)
```

```python
# b. Xác suất để số chấm xuất hiện lớn hơn hoặc bằng 4
rolls4_6=np.count_nonzero(rolls>=4)/num_rolls
print("Xac suat ra mat 4 hoac 5 hoac 6 la: ",rolls4_6)
```

```python
# c. Giả sử biết số chấm xuất hiện lớn hơn hoặc bằng 4. Hãy tìm xác suất để mặt 6 chấm xuất hiện.
rolls6_4=rolls[rolls>=4]
rolls6_4_6=np.count_nonzero(rolls6_4==6)/len(rolls6_4)
print("Xac suat ra mat 6 khi biet ra mat 4 hoac 5 hoac 6 la: ",rolls6_4_6)
```

**Bài 05:** Tạo 5000 số ngẫu nhiên có phân phối nhị thức với n=50, p=0.7 Bạn Hãy

a. Tìm các giá trị thống kê: min, max, Q1, Q2, Q3

b. Vẽ đồ thị boxplot

c. Vẽ các đồ thị: tần số, tần suất

```python
n=50
p=0.7
size=5000
dt=np.random.binomial(n,p,size)

# a) Cac gia tri thong ke min,max,Q1,Q2,Q3
print("Gia tri nho nhat la: ",np.min(dt))
print("Gia tri lon nhat la: ",np.max(dt))
print("Gia tri Q1 la: ",np.percentile(dt,25))   
print("Gia tri Q2 la: ",np.percentile(dt,50))
print("Gia tri Q3 la: ",np.percentile(dt,75))
```

```python
# b) Ve do thi boxplot
plt.boxplot(dt)
plt.title("Boxplot of Binomial Distribution (n=50, p=0.7)")
plt.show()
```

```python
# c) ve cac do thi tan so

plt.hist(dt, bins=50)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Binomial Distribution (n=50, p=0.7)")
plt.show()
```

```python
# c) ve cac do thi tan suat

x=np.arange(20)
pmf=binom.pmf(x,n,p)
plt.bar(x,pmf)
plt.xlabel("Value")
plt.ylabel("Probability")
plt.title("PMF of Binomial Distribution (n=50, p=0.7)")
plt.show()
```
